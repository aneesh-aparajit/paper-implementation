```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

## U-Net: Convolutional Networks for Biomedical Image Segmentation

Link: https://arxiv.org/abs/1505.04597

![U-Net](https://lmb.informatik.uni-freiburg.de/people/ronneber/u-net/u-net-architecture.png)

```python
def double_conv(in_channels, out_channels):
    conv = nn.Sequential(
        nn.Conv2d(in_channels=in_channels, out_channels=out_channels, 
                  kernel_size=(3, 3), stride=(1, 1), padding=(0, 0)), 
        nn.BatchNorm2d(num_features=out_channels),
        nn.ReLU(), 
        nn.Conv2d(in_channels=out_channels, out_channels=out_channels, 
                  kernel_size=(3, 3), stride=(1, 1), padding=(0, 0)), 
        nn.BatchNorm2d(num_features=out_channels), 
        nn.ReLU()
    )
    return conv
```

```python
def crop_image(tensor, target_size):
    """
    Extracted from: https://github.com/ishandutta0098/paper-implementations/blob/main/Architectures/UNet/unet.ipynb
    """
    tensor_size = tensor.size()[2]
    delta = tensor_size - target_size[1]
    delta = delta // 2
    return tensor[:, :, delta:tensor_size-delta, delta:tensor_size-delta]
```

```python
class UNet(nn.Module):
    def __init__(self, in_channels):
        """
        Expected Input: (None, 1, 572, 572)
        """
        super(UNet, self).__init__()
        # down_conv
        self.down_conv1 = double_conv(in_channels=in_channels, out_channels=64)
        self.down_conv2 = double_conv(in_channels=64, out_channels=128)
        self.down_conv3 = double_conv(in_channels=128, out_channels=256)
        self.down_conv4 = double_conv(in_channels=256, out_channels=512)
        self.down_conv5 = double_conv(in_channels=512, out_channels=1024)
        
        # up_conv
        self.up_conv1_trans = nn.ConvTranspose2d(in_channels=1024, out_channels=512, 
                                                 kernel_size=(2, 2), stride=(2, 2))
        self.up_conv1 = double_conv(in_channels=1024, out_channels=512)
        
        self.up_conv2_trans = nn.ConvTranspose2d(in_channels=512, out_channels=256, 
                                                 kernel_size=(2, 2), stride=(2, 2))
        self.up_conv2 = double_conv(in_channels=512, out_channels=256)
        
        self.up_conv3_trans = nn.ConvTranspose2d(in_channels=256, out_channels=128, 
                                                 kernel_size=(2, 2), stride=(2, 2))
        self.up_conv3 = double_conv(in_channels=256, out_channels=128)
        
        self.up_conv4_trans = nn.ConvTranspose2d(in_channels=128, out_channels=64, 
                                                 kernel_size=(2, 2), stride=(2, 2))
        self.up_conv4 = double_conv(in_channels=128, out_channels=64)
        
        self.up_conv5 = nn.Conv2d(in_channels=64, out_channels=2, kernel_size=1)
        
        self.maxpool = nn.MaxPool2d(kernel_size=(2, 2), stride=(2, 2))
    
    def forward(self, x):
        print(f'x -> {x.shape}')
        down_conv_results = []
        
        x = self.down_conv1(x)
        down_conv_results.append(x)
        print(f'x -> {x.shape}')
        x = self.maxpool(x)
        print(f'x -> {x.shape}')
        
        x = self.down_conv2(x)
        down_conv_results.append(x)
        print(f'x -> {x.shape}')
        x = self.maxpool(x)
        print(f'x -> {x.shape}')
        
        x = self.down_conv3(x)
        down_conv_results.append(x)
        print(f'x -> {x.shape}')
        x = self.maxpool(x)
        print(f'x -> {x.shape}')
        
        x = self.down_conv4(x)
        down_conv_results.append(x)
        print(f'x -> {x.shape}')
        x = self.maxpool(x)
        print(f'x -> {x.shape}')
        
        x = self.down_conv5(x)
        print(f'x -> {x.shape}')
        
        x = self.up_conv1_trans(x)
        print(f'x -> {x.shape}')
        x = self.up_conv1(torch.cat([x, 
                                     crop_image(down_conv_results[3], x.shape[2:])], dim=1))
        print(f'x -> {x.shape}')
        
        x = self.up_conv2_trans(x)
        print(f'x -> {x.shape}')
        x = self.up_conv2(torch.cat([x, 
                                     crop_image(down_conv_results[2], x.shape[2:])], dim=1))
        print(f'x -> {x.shape}')
        
        x = self.up_conv3_trans(x)
        print(f'x -> {x.shape}')
        x = self.up_conv3(torch.cat([x, 
                                     crop_image(down_conv_results[1], x.shape[2:])], dim=1))
        print(f'x -> {x.shape}')
        
        x = self.up_conv4_trans(x)
        print(f'x -> {x.shape}')
        x = self.up_conv4(torch.cat([x,
                                     crop_image(down_conv_results[0], x.shape[2:])], dim=1))
        print(f'x -> {x.shape}')
        
        x = self.up_conv5(x)
        print(f'x -> {x.shape}')
        return x
```

```python
X = torch.rand(size=(1, 1, 572, 572))
model = UNet(in_channels=1)
y = model.forward(X)
```

```python
plt.figure(figsize=(18, 24))

plt.subplot(131)
plt.imshow(X[0].permute([2, 1, 0]), cmap='gray')
plt.title('X')

plt.subplot(132)
plt.imshow(y[0][0].detach().numpy(), cmap='gray')
plt.title('y[0][0]')

plt.subplot(133)
plt.imshow(y[0][1].detach().numpy(), cmap='gray')
plt.title('y[0][1]')

plt.tight_layout()
```
